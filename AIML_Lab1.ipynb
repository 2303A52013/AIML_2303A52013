**Part-02:  Visualization of Data using matplotlib, pyplots Pack
ages**


```python
import matplotlib.pyplot as plt
```

```python
people = ['Deepak', 'Bharath', 'John', 'Tom', 'Nawaz']
age = [22, 31, 55, 44, 18]
weight = [55, 62, 80, 68, 75]
height = [150, 142, 172, 155, 167]
```

**Scatterplot**

```python
plt.scatter(weight, height)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Relationship between Height and Weight of Patiants')
plt.show()
```

**Bar Chart**

```python
plt.bar(people, height)
plt.xlabel('People')
plt.ylabel('Height')
plt.title('Height of Patiants')
plt.show()
```

**Histogram**

```python
plt.hist(age, bins=6, color='green', edgecolor ='black')
plt.xlabel('Age')
plt.ylabel('No. of Patients')
plt.title('Age of Patients')
plt.show()
```
